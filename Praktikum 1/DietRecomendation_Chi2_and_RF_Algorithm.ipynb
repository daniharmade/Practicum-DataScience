```python
import matplotlib.pyplot as plt
import seaborn as sns

# fungsi untuk plot distribusi data
def plot_distribution(data, x_axis, y_axis, hue_col=None, title=''):
    plt.figure(figsize=(8, 5))
    c1 = sns.barplot(x=x_axis, y=y_axis, data=data,
                     palette='viridis', hue=hue_col,
                     legend=False)
    for container in c1.containers:
        c1.bar_label(container)
    plt.title(title)
    plt.show()
```

```python
def with_per(total, ax):
    for container in ax.containers:
        # mengambil semua bar (rects) dalam plot
        rects = container.get_children()
        for rect in rects:
            height = rect.get_height()
            percentage = '{:.1f}%'.format(100 * height / total)

            # menentukan posisi untuk label persentase
            x = rect.get_x() + rect.get_width() / 2
            y = height

            # menambahkan anotasi di atas setiap bar
            ax.annotate(percentage, (x, y), ha='center', va='center',
                        xytext=(0, 8), textcoords='offset points', fontsize=12)
```

```python
# cek data outlier
import numpy as np
import pandas as pd

def detect_outliers_iqr(df):
    df_numeric = df.select_dtypes(include=[np.number])
    Q1 = df_numeric.quantile(0.25)
    Q3 = df_numeric.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return (df_numeric < lower_bound) | (df_numeric > upper_bound)

outliers = detect_outliers_iqr(df_clean)
outliers.sum()
```

```python
data = df_clean.copy()
```

```python
# mengubah tipe data kategorikal menjadi representasi numerik
from sklearn.preprocessing import LabelEncoder

data_fe = data.copy()

object_cols = [col for col in data_fe.columns if data_fe[col].dtype == 'object']

label_encoder = LabelEncoder()
for col in object_cols:
    data_fe[col] = label_encoder.fit_transform(data_fe[col])

data_fe.tail()
```

```python
# memisahkan fitur dengan label
X = data_fe.drop(columns='Diet_Recommendation')
X = X.fillna(X.median())
y = data_fe['Diet_Recommendation']
```

```python
# implemetasi Chi2 sebagai seleksi fitur
from sklearn.feature_selection import chi2

chi_square = chi2(X, y)

# menampilkan Chi-Score
chi_score = pd.DataFrame({'Feature': X.columns, 'Chi_Score': chi_square[0]})
chi_score = chi_score.sort_values(by='Chi_Score', ascending=False)
chi_score
```

```python
# visualisasi Chi Score
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.bar(chi_score['Feature'], chi_score['Chi_Score'])
plt.xlabel('Features')
plt.ylabel('Chi-Square Score')
plt.title('Nilai Chi Score pada Setiap Fitur')
plt.xticks(rotation=90)
plt.show()
```

Berdasarkan visualisasi nilai Chi-Square, fitur Disease_Type memiliki kontribusi paling signifikan terhadap penentuan rekomendasi diet, diikuti oleh Daily_Caloric_Intake dengan skor yang jauh lebih rendah namun tetap mencolok dibanding fitur lainnya. Hal ini menunjukkan bahwa jenis penyakit dan asupan kalori harian merupakan faktor utama yang memengaruhi keputusan dalam pemberian rekomendasi diet. Sementara itu, fitur-fitur lain seperti Height_cm, Weekly_Exercise_Hours, dan Age memiliki pengaruh yang jauh lebih kecil, dan fitur seperti Dietary_Nutrient_Imbalance_Score, Severity, dan Preferred_Cuisine hampir tidak memberikan kontribusi berarti dalam model klasifikasi.

```python
# menampilkan P-Values
p_values = pd.DataFrame({'Feature': X.columns, 'P-Values': chi_square[1]})
p_values = p_values.sort_values(by='P-Values', ascending=True)
p_values
```

```python
# visualisasi p-value
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.bar(p_values['Feature'], p_values['P-Values'])
plt.xlabel('Features')
plt.ylabel('Chi-Square P-Values')
plt.title('Nilai P-Values pada Setiap Fitur')
plt.xticks(rotation=90)
plt.show()
```

Berdasarkan grafik nilai p-value, fitur Disease_Type dan Daily_Caloric_Intake menunjukkan nilai p yang sangat rendah, menandakan bahwa keduanya memiliki hubungan yang sangat signifikan terhadap output rekomendasi diet. Semakin tinggi nilai p, semakin kecil signifikansi fitur tersebut, sehingga fitur seperti Severity, Preferred_Cuisine, dan Dietary_Nutrient_Imbalance_Score memiliki kontribusi yang tidak signifikan dalam menentukan rekomendasi diet. Dengan demikian, fitur yang memiliki p-value kecil layak diprioritaskan dalam pengembangan model prediktif, sementara fitur dengan p-value tinggi dapat dipertimbangkan untuk dikurangi atau dihilangkan dari model.

```python
# ambil fitur penting/memiliki nilai p-value < 0.05
selected_feature = p_values[p_values['P-Values'] < 0.05]
selected_feature
```

```python
# seleksi fitur pada dataset
X_fs = X[selected_feature['Feature'].tolist()]
X_fs
```

```python
# normalisasi dengan Min Max Normalization
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0, 1))
X_fs_sc = sc.fit_transform(X_fs)
```

# Pemodelan

```python
# pembagian data dengan Hold-Out Validation
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_fs_sc, y, test_size=0.3, random_state=42)
```

```python
# import matrik evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
```

```python
# implementasi algoritma RF
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    'n_estimators': [10, 50, 100, 200],  # masukkan jumlah pohon (estimators)
    'max_depth': [10, 20, 30],  # masukkan kedalaman maksimum
    'min_samples_leaf': [1, 2, 4]  # Jumlah sampel minimum yang diperlukan untuk menjadi node leaf
}

# inisialisasi model RF
model_rf = RandomForestClassifier(random_state=42)

# inisialisasi GridSearchCV
grid_search = GridSearchCV(model_rf, param_grid, cv=5, scoring='accuracy')

# melatih model dengan Grid Search
grid_search.fit(X_train, y_train)

# mendapatkan parameter terbaik
best_params = grid_search.best_params_

# mendapatkan model terbaik
best_model = grid_search.best_estimator_

print(best_params)
```

```python
# implementasi RF dengan parameter terbaik
best_model.fit(X_train, y_train)

y_pred_rf = best_model.predict(X_test)
print(classification_report(y_test, y_pred_rf))
```

```python
# membuat fungsi untuk menampilkan confusion matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", linewidths=0.5, linecolor="black")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()
```

```python
# menampilkan confusion matrix untuk model RF
plot_confusion_matrix(y_test, y_pred_rf, "Confusion Matrix (RF)")
```

Berdasarkan confusion matrix model Random Forest (RF) di atas, dapat dilihat bahwa model mampu mengklasifikasikan data dengan cukup baik, khususnya pada kelas 1 yang seluruh prediksinya tepat (67 benar, tanpa kesalahan). Namun, terdapat kesalahan klasifikasi yang cukup signifikan pada kelas 0 dan kelas 2, di mana 35 data dari kelas 0 diklasifikasikan sebagai kelas 2, dan 33 data dari kelas 2 diklasifikasikan sebagai kelas 0. Hal ini menunjukkan adanya kebingungan antara kelas 0 dan 2, yang bisa disebabkan oleh kemiripan karakteristik fitur antara kedua kelas tersebut. Meskipun begitu, performa model secara keseluruhan tetap cukup baik, terutama karena akurasi prediksi pada masing-masing kelas tidak terlalu timpang.
